```python
import csv
from collections import defaultdict 
```

```python
graph = defaultdict(list) 
def addEdge(graph,u,v): 
    graph[u].append(v) 
```

Defining an the DFS graph seen checks if a node is visted or not path adds the nodes to the path visit tells how many times a node is visited.

```python
mydict={}
def DFS(G,v,seen=None,path=None,visited=None):
    if seen is None: seen = []
    if path is None: path = [v]
    if visited is None: visited={}

    cin=visited.get(v)
    if cin:
        visited[v]=cin+1
    else:
        visited[v]=1
        
    seen.append(v)

    paths = []
    for t in G[v]:
        var1=visited.get(t)
        var2=mydict.get(t)
        if var1!=var2 or t not in seen:
            t_path = path + [t]
            paths.append(tuple(t_path))
            paths.extend(DFS(G, t, seen[:], t_path,visited))
    return paths
```
